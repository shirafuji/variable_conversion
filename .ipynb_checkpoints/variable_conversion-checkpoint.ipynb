対数差分変換の説明用プログラム

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
passengers = pd.read_csv('AirPassengers.csv', index_col=0, parse_dates=[0])
plt.plot(passengers)
print(passengers)
```

```python
# 対数変換
# logを取ることによってトレンドに比例したばらつきの増加が解消される
passengers_log = np.log(passengers.values)
plt.plot(passengers_log)
```

```python
# 差分変換
# 一時点離れた値との差を取る差分変換により平均に関して非定常な系列を定常にできる場合がある
passengers_log_diff = []
for i in range(len(passengers_log)-1):
    passengers_log_diff.append(passengers_log[i+1]-passengers_log[i])
plt.plot(passengers_log_diff)
```

このような一連の流れを対数差分変換と呼ぶ
