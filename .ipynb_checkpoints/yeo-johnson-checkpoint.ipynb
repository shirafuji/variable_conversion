yeo-johnso変換の説明用サンプルコード

```python
import numpy as np

def yeo_johnson(x, alpha):
    y = []
    for i in range(len(x)):
        if x[i] >= 0:
            if alpha != 0:
                y.append(((x[i]+1)**alpha-1)/alpha)
            else:
                y.append(np.log(x[i]+1))
        else:
            if alpha != 2:
                y.append(-((-x[i]+1)**(2-alpha)-1)/(2-alpha))
            else:
                y.append(-np.log(-x[i]+1))
    return y
```

```python
x = np.arange(-2.9, 3.1, 0.1)
alpha = -1
y = []
for i in range(6):
    y.append(yeo_johnson(x, alpha))
    alpha += 0.5
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.set_title("yeo-johnson")
ax.set_xlabel('x')
ax.set_ylabel('y')
c1,c2,c3,c4,c5 = "blue","green","red","black","yellow"      # 各プロットの色
l1,l2,l3,l4,l5 = "alpha=-1","alpha=-0.5","alpha=0","alpha=0.5","alpha=1"   # 各ラベル
ax.plot(x, y[0], color=c1, label=l1)
ax.plot(x, y[1], color=c2, label=l2)
ax.plot(x, y[2], color=c3, label=l3)
ax.plot(x, y[3], color=c4, label=l4)
ax.plot(x, y[4], color=c5, label=l5)
plt.show()
```

```python
from scipy import stats
y1 = np.random.exponential(scale=0.5, size=1000000) - 0.5
y1_lambda, lmax1 = stats.yeojohnson(y1)
y2 = np.random.beta(2.0, 5.0, size=1000000) - 3
y2_lambda, lmax2 = stats.yeojohnson(y2)
bins = 100
plt.rcParams["font.size"] = 15
plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1)
plt.hist(y1, bins=bins)
plt.title(r'$y_1$')
plt.subplot(2, 2, 2)
plt.hist(y1_lambda, bins=bins)
plt.title(r'$y_1^{(\lambda)} (\lambda=%f)$' % lmax1)
plt.subplot(2, 2, 3)
plt.hist(y2, bins=bins)
plt.title(r'$y_2$')
plt.subplot(2, 2, 4)
plt.hist(y2_lambda, bins=bins)
plt.title(r'$y_2^{(\lambda)} (\lambda=%f)$' % lmax2)
plt.tight_layout()
plt.savefig('figure.png')
plt.show()
```
